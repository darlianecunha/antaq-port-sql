# antaq-port-sql — exploration

Reproducible SQL warehouse of Brazilian port operations (ANTAQ-style). Sample data is synthetic.

First build the warehouse: `python load.py` (or run the cell below).

```
import duckdb, pandas as pd
# build if needed
import subprocess, os
if not os.path.exists('../antaq.duckdb'):
    subprocess.run(['python','load.py'], cwd='..', check=True)
con = duckdb.connect('../antaq.duckdb', read_only=True)
```

## Top ports by cargo (2024)

```
con.execute(open('../queries/01_top_ports_by_cargo.sql').read()).fetchdf()
```

## Year-over-year growth (LAG window)

```
yoy = con.execute(open('../queries/03_yoy_growth.sql').read()).fetchdf()
yoy
```

## Cargo nature mix per port

```
mix = con.execute(open('../queries/04_cargo_nature_mix.sql').read()).fetchdf()
mix.set_index('port_name')[['pct_solid_bulk','pct_liquid_bulk','pct_container','pct_general']].plot(kind='bar', stacked=True, title='Cargo nature mix by port (%)')
```
